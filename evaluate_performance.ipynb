```python
import pandas as pd 
from helper import *

path = 'performance'
df = load_data(f'{path}/driving_log.csv')
df
```

```python
img_path, steering = load_data_to_array(path, df)
img = img_preprocess_pipeline(img_path, steering, train_flag=False)
```

```python
from keras.models import load_model
from keras.metrics import mean_squared_error
import numpy as np

model_name = 'model.h5'
model = load_model(f'model/{model_name}')
pred_val = []
for i in img[0]:
    image = np.asarray([i])
    steering_val = float(model.predict(image, verbose=2, batch_size=1))
    pred_val.append(steering_val)
df = pre_perf(df, pred_val)
pred_val
```

```python
mse = mean_squared_error(pred_val, df['Steering'])
mse
```

```python
import plotly.express as px 
import plotly.graph_objects as go 

fig = px.line(df, y=['Steering', 'Prediction'])
fig.show()
```
