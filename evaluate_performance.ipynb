```python
import pandas as pd 
from helper import *

path = 'performance'
df = load_data(f'{path}/driving_log.csv')
```

```python
img_path, steering = load_data_to_array(path, df)
img = img_preprocess_pipeline(img_path, steering, train_flag=False)
```

```python
from keras.models import load_model
from keras.metrics import mean_squared_error
import numpy as np

model_name ='Adam-0.00505.hdf5'
model = load_model(f'model/{model_name}')
pred_val = []
for i in img[0]:
    image = np.asarray([i])
    steering_val = float(model.predict(image, verbose=2, batch_size=1))
    pred_val.append(steering_val)
```

```python
def fwhm2sigma(fwhm):
    return fwhm / np.sqrt(8 * np.log(2))
```

```python
from helper import * 

df['Prediction'] = pred_val
window_size = 70
std = fwhm2sigma(window_size)
df = df.rolling(window=window_size, win_type='gaussian', center=True).mean(std=std).dropna()
# df = pre_perf(df)
```

```python
import plotly.express as px 
import plotly.graph_objects as go 

fig = px.line(df, y=['Prediction','Steering'], 
            title='Performance evaluation',
            labels={'value':'Steering values', 'index': 'Time'},
            color_discrete_map={'Prediction':'red', 'Steering':'blue'})
# fig.update_layout(yaxis_range=[-0.4, 0.4])
fig.show()
```
